## Funções disponíveis
| Função | Descrição |
|---|---|
| `nulos_por_coluna(df, nome)` | Exibe a quantidade de valores nulos por coluna de um DataFrame |

````python
from pyspark.sql.functions import col, sum as spark_sum

def nulos_por_coluna(df, nome):
    print(f"=== NULOS - {nome} ===")
    display(
        df.select([
            spark_sum(col(c).isNull().cast("int")).alias(c)
            for c in df.columns
        ])
    )
````
