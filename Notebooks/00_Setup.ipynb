# 00_Setup - Bike Store Medallion
Notebook de configuração inicial do projeto.

Responsável por criar toda a infraestrutura necessária no Unity Catalog:
- **Catálogo:** `bike_store`
- **Schemas:** `landing`, `bronze`, `silver`, `gold`
- **Volume:** `bike_store/landing/raw_files`

> ⚠️ Este notebook deve ser executado **apenas uma vez** antes de iniciar o projeto.

```python
# Criar o catálogo
spark.sql("CREATE CATALOG IF NOT EXISTS bike_store")
```

```python
# Criar os schemas
spark.sql("CREATE SCHEMA IF NOT EXISTS bike_store.landing")
spark.sql("CREATE SCHEMA IF NOT EXISTS bike_store.bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS bike_store.silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS bike_store.gold")
```

```python
# Criar o volume para receber os CSVs
spark.sql("CREATE VOLUME IF NOT EXISTS bike_store.landing.raw_files")
```

```python
# Verificar o que foi criado
print("=== CATÁLOGO ===")
display(spark.sql("SHOW CATALOGS LIKE 'bike_store'"))

print("=== SCHEMAS ===")
display(spark.sql("SHOW SCHEMAS IN bike_store"))

print("=== VOLUMES ===")
display(spark.sql("SHOW VOLUMES IN bike_store.landing"))
```
