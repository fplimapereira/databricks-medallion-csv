# 01 - Ingestão Bronze - Bike Store Medallion
Leitura dos arquivos CSV do Volume `bike_store/landing/raw_files`
e gravação como Delta Tables na camada `bike_store.bronze`.

> Por ser um projeto mais simples de estudos, as tabelas são criadas com schema inferido automaticamente.

```python
# Caminho do volume com os CSVs
volume_path = "/Volumes/bike_store/landing/raw_files"

# Catálogo e schema de destino
catalog = "bike_store"
schema  = "bronze"

# Arquivos a ingerir
tabelas = [
    "brands",
    "categories",
    "customers",
    "order_items",
    "orders",
    "products",
    "staffs",
    "stocks",
    "stores"
]
```

```python
# Montando o caminho completo para cada arquivo csv e apontando para seu destino 
for tabela in tabelas:
    caminho_csv = f"{volume_path}/{tabela}.csv"
    destino     = f"{catalog}.{schema}.{tabela}"

    df = spark.read.csv(
        caminho_csv,
        header=True,
        inferSchema=True
    )

    df.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable(destino)

    print(f"✅ {destino} — {df.count()} linhas")
```
