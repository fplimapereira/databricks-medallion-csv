# 99 - Validação Silver - Bike Store Medallion
Notebook de validação da camada Silver.

Verifica se todas as tabelas Silver foram criadas e possuem registros.

Tabelas validadas:
- `bike_store.silver.products`
- `bike_store.silver.orders`
- `bike_store.silver.clients`

```python
tabelas = [
    "bike_store.silver.products",
    "bike_store.silver.orders",
    "bike_store.silver.clients"
]

erros = []

for tabela in tabelas:
    try:
        count = spark.read.table(tabela).count()
        if count == 0:
            erros.append(f"❌ {tabela} — tabela vazia!")
        else:
            print(f"✅ {tabela} — {count} registros")
    except Exception as e:
        erros.append(f"❌ {tabela} — tabela não encontrada!")

if erros:
    for erro in erros:
        print(erro)
    raise Exception("❌ Validação falhou! Verifique as tabelas acima.")
else:
    print("\n✅ Todas as tabelas Silver validadas com sucesso!")
```
